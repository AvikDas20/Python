List all customers' full names and their email addresses.

```python
import pandas as pd

df = pd.read_csv('customer.csv')
df[['first_name','last_name', 'email']]
```

Retrieve all film titles with their rental durations and rental rates.

```python
film_df = pd.read_csv('film.csv')
film_df[['title','rental_duration','rental_rate']]
```

Display the customer’s email, their address ID, and the name of the city they live in.

```python
customer_df = pd.read_csv('customer.csv')
city_df = pd.read_csv('city.csv')
address_df = pd.read_csv('address.csv')

city_address_df = pd.merge(city_df,address_df,how='left',on='city_id')
customer_city_address_df = pd.merge(customer_df,city_address_df,how='left',on='address_id')
customer_city_address_df[['customer_id','email','address_id','city']]
```

Find the titles of all films that are currently not in inventory.

```python
inventory_df = pd.read_csv('inventory.csv')

print(inventory_df.shape)
print(film_df.shape)
films_inventory_missing = pd.merge(film_df,inventory_df,how='left',on='film_id')
films_not_in_inventory = films_inventory_missing[films_inventory_missing['inventory_id'].isna()]
result = films_not_in_inventory[['title']].drop_duplicates()
print(result)
```

Show the top 5 films with the longest rental duration.

```python
final_df  = film_df[['title', 'rental_duration']]
final_df.sort_values('rental_duration', ascending=False, inplace=True)
final_df.head(5)
```

Count how many customers are from each country.

```python
country_df = pd.read_csv('country.csv')
country_city_df = pd.merge(country_df,city_df,how='left',on='country_id').drop(columns=['country_id','last_update_x','last_update_y','city'])
country_city_address_df = pd.merge(address_df,country_city_df,how='left',on='city_id').drop(columns=['address','address2','last_update','district','postal_code','phone'])
customer_country_city_address_df = pd.merge(customer_df,country_city_address_df,how='left',on='address_id')
customer_counts = customer_country_city_address_df.groupby(by='country')['customer_id'].count().reset_index()
customer_counts.columns = ['country', 'customer_count']
customer_counts.sort_values('customer_count', ascending=False, inplace=True)
customer_counts
```
